```python
df=pd.DataFrame(np.random.randn(100,3),columns=['A','B','C'])
df['A']=df['A'].cumsum()+10
df['B']=df['B'].cumsum()+10
df['C']=df['C'].cumsum()+10
```

```python
df.iplot()
```

```python
df.iplot(kind='area',fill=True)
```
