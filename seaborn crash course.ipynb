## controlling plotted figure aesthetics
fig styling
axes styling
color palattes etc

```python
def sinplot(flip=1):
    x=np.linspace(0,14,100)
    print(x)
    for i in range(1,7):
        plt.plot(x,np.sin(x+i*0.5)*(7-i)*flip)
```

```python
sinplot()
```

```python
sns.set_style('white') #white whitegrid ticks dark
sinplot()
```

```python
sns.set_style('white') #white whitegrid ticks dark
sinplot()
sns.despine()
```

```python
sns.set_style('white') #white whitegrid ticks dark
sinplot()
sns.despine(left=True)
```

```python
sns.set_style('white') #white whitegrid ticks dark
sinplot()
sns.despine(left=True,bottom=True)
```

```python
sns.axes_style()
```

```python
sns.set_style('ticks',{'axes.grid':True}) #white whitegrid ticks dark
sinplot()
sns.despine(left=True)
```

```python
sns.set_style('ticks',{'axes.grid':True,'xtick.direction':'in'}) #white whitegrid ticks dark
sinplot()
sns.despine(left=True)
```
